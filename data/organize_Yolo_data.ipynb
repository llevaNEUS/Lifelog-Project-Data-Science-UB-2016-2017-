```python
import pandas as pd

csv_file_merge_u2_u3 = "/Users/Aneta/Desktop/Final_Project_Aneta/merge_u2_u3.csv"
merge_u2_u3= pd.read_csv(csv_file_merge_u2_u3, sep=',', header=None)
merge_u2_u3 = merge_u2_u3.iloc[:, :-48]
merge_u2_u3.columns=columns

merge_u2_u3['user_yolo'] = merge_u2_u3.path.str.split('/').str.get(1)
merge_u2_u3['user_yolo'].unique()

```

```python
data_u2 = merge_u2_u3[merge_u2_u3['user_yolo'] == "u2"]
data_u2

df_u2_unique=data_u2[['path']]
df_u2_unique["path_len"]=df_u2_unique['path'].str.len()
df_u2_unique["path_len"].value_counts()
```

```python
to_drop_u2 = df_u2_unique["path_len"] > 52.0
to_drop_u2_done=df_u2_unique[to_drop_u2]
list_drop_u2=to_drop_u2_done.index.tolist()
data_u2_clean=data_u2.drop(data_u2.index[list_drop_u2])
data_u2_clean["path_len"]=data_u2_clean['path'].str.len()
data_u2_clean["path_len"].value_counts()
data_u2_clean
```

```python
data_u2_clean['path_id'] = data_u2_clean["path"].str.split('/').str.get(3)
data_u2_clean['path_id'] = data_u2_clean['path_id'].str[:-4]
data_u2_clean['path_id'].unique()

y = data_u2_clean['path_id']
data_u2_clean.drop(labels=['path_id'], axis=1,inplace = True)
data_u2_clean.insert(0, 'path_id', y)

x = data_u2_clean['user_yolo']
data_u2_clean.drop(labels=['user_yolo'], axis=1,inplace = True)
data_u2_clean.insert(0, 'user_yolo', x)

data_u2_clean= data_u2_clean.drop(["path", "path_len"], axis=1)

data_u2_clean
```

```python
data_u3 = merge_u2_u3[merge_u2_u3['user_yolo'] == "u3"]

df_u3_unique=data_u3[['path']]
df_u3_unique["path_len"]=df_u3_unique['path'].str.len()
df_u3_unique["path_len"].value_counts()
data_u3_clean=data_u3

data_u3_clean['path_id'] = data_u3_clean.path.str.split('/').str.get(4)
data_u3_clean['path_id'] = data_u3_clean['path_id'].str[:-4]
data_u3_clean['path_id'].unique()

y = data_u3_clean['path_id']
data_u3_clean.drop(labels=['path_id'], axis=1,inplace = True)
data_u3_clean.insert(0, 'path_id', y)

x = data_u3_clean['user_yolo']
data_u3_clean.drop(labels=['user_yolo'], axis=1,inplace = True)
data_u3_clean.insert(0, 'user_yolo', x)

data_u3_clean= data_u3_clean.drop(["path"], axis=1)
data_u3_clean
```
